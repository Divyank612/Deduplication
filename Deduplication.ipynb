### Exact duplicate - Index of the rows

```python
def check_duplicate(df):
    check_list = df[df.duplicated()].index
    if check_list.empty:
        print("no duplicates")
    else:
        print("duplicate found")
        return check_list
```

```python
duplicate_index = list(check_duplicate(data))
print(duplicate_index)
```

```python
## REMOVING THE EXACT DUPLICATE ROWS
data = data.drop_duplicates(keep='first')
```

```python
data
```

```python
check_duplicate(data)
```

```python
data['is_partial_duplicate'] = data.duplicated(subset=['user_id','game_title'], keep=False)
print(data)
```
